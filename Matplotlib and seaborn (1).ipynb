#How to make scatter plot

```python
sns.scatterplot(data=df, x='Runs',y='Price',hue='Age')
plt.title('Player Experience vs Price')
plt.show()
```

```python
top_players= df.sort_values(by='Price',ascending= False)
top_players.head(5)
```

#making of correlation heatmap


```python
numeric_df = df.select_dtypes(include=['float64','int64'])
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('What stats correlate with Price?')
plt.show()
```

#Making Histogram

```python
plt.hist(df["Price"],bins=20,color='maroon')
plt.title("Distribution of Player Prices",color='blue')
plt.xlabel("Price (INR Lakhs)",color='red')
plt.ylabel("Number of Players")
plt.show()
```

#Making Bar Chart

```python
price_exp=df.groupby("Experience_Years")["Price"].mean()
price_exp.plot(kind="bar",color='Red')
plt.title("Does Experience drive player price? - Average Price by experience years")
plt.xlabel("Years of Professional experience",color="Red")
plt.ylabel("Average price of player ", color='pink')
plt.xticks(rotation=35)
plt.show ()
```

#Creating a Pie Chart

```python
df["Age_Group"] = pd.cut(df["Age"], bins=[18, 25, 30, 36], labels=["Young", "Mid", "Senior"])
age_counts = df["Age_Group"].value_counts()
plt.pie(age_counts, labels=age_counts.index, autopct="%1.2f%%")
plt.title("Composition of Player Pool by Career Age Group")
plt.show()
```

#Counting Players through Charts

```python
sns.countplot(x="Age_Group", data=df)
plt.title("Number of players in each career age group")
plt.xlabel("Career age group")
plt.ylabel("Number of players")
plt.show()
```

#Creating a Boxplot

```python
sns.boxplot(x=df["Price"])
plt.title("Distribution of player prices- Identifying Outliers")
plt.xlabel("Player Price(Inr Lakhs)")
plt.show()
```

```python
sns.regplot(x='Runs', y='Price', data=df)
plt.title('Runs vs Price')
plt.xlabel('Career Runs Scored')
plt.ylabel('Player Price (INR Lakhs)')
plt.show()  
```

```python
numeric_df = df.select_dtypes(include=[np.number])
corr_matrix = numeric_df.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f") 
plt.title("Correlation Matrix")
plt.show()  
```

```python
sns.violinplot(x="Age_Group", y="Price", data=df)
plt.title("Price Distribution by Career Age Group,fontsize=1")
plt.xlabel("Career Age Group")
plt.ylabel("Player Price (INR Lakhs)")
plt.show()
```
