MAKING BMI

```python
#ht=height,wt=weight
std_height=1.76
std_weight=68.7
```

```python
print ("Height=", std_height)
print ("Weight=", std_weight)
```

```python
#Calculating Body Mass Index
BMI=std_weight/std_height**2
print("BMI=", round(BMI,2))
```

```python
type (BMI)
```

PYTHON MATH LIBRARY

```python
#importing the math library
import math
```

```python
pi_value =math.pi
print(pi_value)
```

```python
# Print value of e
e_value = math.e
print(e_value)
```

```python
# other math functions
# Find 8 to the power 2
pow = math.pow(8,2)
print("8 to the power 2 is:", pow)
# What is square root of 64.
sqrt = math.sqrt(64)
print("square root of 64 is:", sqrt)
# Find factorial of 4
fact = math.factorial(4)
print("factorial of 4 is:", fact)
```

```python
# Find area of a circle
r = int(input("Enter the value of radius:"))
```

```python
pi = math.pi
print("r =", r)
print("pi =", round(pi,2))
```

```python
area = pi * r * r
print("Area of Circle is:", round(area,2))
```
