Scatter plot

```python
sns.scatterplot(data=df,x='Performance_Score',y='Age',hue='Gender')
plt.title('Performance_Score vs Age')
plt.show()
```

### Corelation Map

```python
numeric_df = df.select_dtypes(include=['float64','int64'])
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('What stats correlate with Performance?')
plt.show()
```

```python
sns.boxplot(x=df["Salary"])
plt.title("Distribution of Salary- Identifying Outliers")
plt.xlabel("Employee Salary(Inr)")
plt.show()
```

### The Salary majorly ranges from 22-45

```python
sns.countplot(x="Department", data=df)
plt.title("Number of Employees in each career")
plt.xlabel("Department")
plt.ylabel("Number of Employee")
plt.show()
```

### Finance and it has the most number of hires

```python
df["Gender"] = df["Gender"].map({"Male": "Male", "Female": "Female"})
age_counts=df["Gender"].value_counts()
plt.pie(age_counts,labels=age_counts.index,autopct="%1.2f%%")
plt.title("No. of employee as gender wise")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x="Department", y="Salary", data=df)

plt.title("Salary Range across Departments")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd  # Added pandas import

# Define your DataFrame first - either by loading from a file
# df = pd.read_csv('your_data_file.csv')  # Uncomment and modify this line to load your data
# Or create a sample DataFrame for testing
df = pd.DataFrame({
    'Department': ['HR', 'IT', 'Sales', 'Marketing', 'Finance', 'Operations'],
    'Salary': [50000, 70000, 60000, 65000, 75000, 55000]
})  # Added DataFrame definition

# 1. Department wise Average Salary nikalna
plt.figure(figsize=(12, 6))

# sns.barplot automatically mean (average) calculate kar leta hai
sns.barplot(x="Department", y="Salary", data=df, palette="viridis")

# 2. Graph ko sundar banana
plt.title("Average Salary by Department", fontsize=15)
plt.xlabel("Department", fontsize=12)
plt.ylabel("Average Salary", fontsize=12)
plt.xticks(rotation=45) # Taaki naam aapas mein takrayein nahi
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

### Here as it is shown that finance has the highest paying job on an average following with it then marketing then sales then operations and hr
